```python
data = pd.get_dummies(data, columns = ['rbc','pc', 'pcc', 'ba','htn', 'dm', 'cad','appet', 'pe', 'ane']) 
#to convert categorical values to numerical values (excluding Nan as they have been already dropped)
```

```python
#encoding the target classification we dont need separate columns for its categorical vales but just encoded in the same column
#That's why label encoding is used
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()

data['classification']= labelencoder_X.fit_transform(data['classification'])
```

```python
dataset = data.select_dtypes(include=['object'])
dataset
```

```python
#still we are left with three columns that are pcv, wc, rc but they contain int and null values
#Therefore, typecasting is done
data = data.convert_objects(convert_numeric=True)
```

```python
y = data['classification']
X = data.drop(['classification'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
dt_scores = []
for i in range(1, len(X.columns) + 1):
    dt_classifier = DecisionTreeClassifier(max_features = i, random_state = 0)
    dt_classifier.fit(X_train, y_train)
    dt_scores.append(dt_classifier.score(X_test, y_test))
```

```python
plt.plot([i for i in range(1, len(X.columns) + 1)], dt_scores, color = 'green')
for i in range(1, len(X.columns) + 1):
    plt.text(i, dt_scores[i-1], (i, dt_scores[i-1]))
plt.xticks([i for i in range(1, len(X.columns) + 1)])
plt.xlabel('Max features')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different number of maximum features')
```

```python
from sklearn.metrics import accuracy_score
print("The score for Decision Tree Classifier is ",dt_scores[17]*100,"%")
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 12, criterion='entropy', random_state =20)
rf.fit(X_train, y_train)

pred_ran=rf.predict(X_test)
accuracy_score(y_test,pred_ran)
```

```python
print("The accuracy score for Random Forest Algorithm is ",accuracy_score(y_test,pred_ran)*100,"%")
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_scores = []
for k in range(1,24):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    knn_classifier.fit(X_train, y_train)
    #knn_scores.append(knn_classifier.score(X_test, y_test))
    pred=knn_classifier.predict(X_test)
    
from sklearn.metrics import accuracy_score
print("The accuracy score of KNN classifier is : ",accuracy_score(y_test,pred)*100,"%")
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print (classification_report(y_test,pred))
print (confusion_matrix(y_test,pred))
```
