```python
data = pd.get_dummies(data, columns = ['rbc', 'pc', 'pcc', 'ba','htn', 'dm', 'cad',
       'appet', 'pe', 'ane'],dummy_na=True) #to convert categorical values to numerical values (including Nan)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
#encoding the target classification we dont need separate columns for its categorical vales but just encoded in the same column
#That's why label encoding is used
data['classification']= labelencoder_X.fit_transform(data['classification'])
```

# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
dt_scores = []
for i in range(1, len(X.columns) + 1):
    dt_classifier = DecisionTreeClassifier(max_features = i, random_state = 0)
    dt_classifier.fit(X_train, y_train)
    dt_scores.append(dt_classifier.score(X_test, y_test))
```

```python
plt.plot([i for i in range(1, len(X.columns) + 1)], dt_scores, color = 'green')
for i in range(1, len(X.columns) + 1):
    plt.text(i, dt_scores[i-1], (i, dt_scores[i-1]))
plt.xticks([i for i in range(1, len(X.columns) + 1)])
plt.xlabel('Max features')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different number of maximum features')
```
